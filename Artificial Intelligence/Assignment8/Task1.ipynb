# Task 1 - Pillow

```python
# Imports
from PIL import Image
from matplotlib import image as Img, pyplot
```

```python
# A. Display an image with pillow:

image = Image.open('spacex.jpg')
print(image.format)
print(image.mode)
print(image.size)
image
```

```python
# B. Convert the image to a numpy array

data = Img.imread('spacex.jpg')
print(data.dtype)
print(data.shape)
pyplot.imshow(data)
pyplot.show()
```

```python
# C. Resize an image to a specific dimension

image = Image.open('spacex.jpg')
print(image.size)
image.thumbnail((100, 100))
print(image.size)
```

```python
# D. Other functions from PIL

# changing image format and saving it
image = Image.open('spacex.jpg')
image.save('spacex.png', format='PNG')
image2 = Image.open('spacex.png')
print(image2.format)

# rotating image
pyplot.subplot(312)
pyplot.imshow(image.rotate(45))

# gray scaling an image
gray_image = image2.convert(mode='L')
gray_image
```
